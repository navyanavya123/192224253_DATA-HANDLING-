```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/superstore.csv",encoding="latin1")

# Extract Product_ID and Product_Name
product_details = df[['Product_ID', 'Product_Name']].drop_duplicates()

# Create numerical indices for products
product_indices = range(len(product_details))

# Plot the data
plt.figure(figsize=(12, 8))
plt.bar(product_indices, len(product_indices) * [1], color='skyblue')
plt.xlabel('Product Index')
plt.ylabel('Count')
plt.title('Product Details')
plt.xticks(product_indices, [])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/superstore.csv",encoding="latin1")

# Group by Product_ID and sum the sales
product_sales = df.groupby('Product_ID')['Sales'].sum()

# Sort products by sales
product_sales_sorted = product_sales.sort_values(ascending=False)

# Plot the top and bottom 10 selling products
top_products = product_sales_sorted.head(10)
bottom_products = product_sales_sorted.tail(10)

plt.figure(figsize=(12, 8))

# Plot top selling products
plt.bar(top_products.index, top_products.values, color='skyblue', label='High-Selling Products')

# Plot low selling products
plt.bar(bottom_products.index, bottom_products.values, color='salmon', label='Low-Selling Products')

plt.xlabel('Product ID')
plt.ylabel('Total Sales')
plt.title('High-Selling and Low-Selling Products by Product ID')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/content/superstore.csv", encoding="latin1")

# Convert 'Order_Date' column to datetime type
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

# Extract year and month from 'Order_Date'
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month

# Group by customer and count the number of purchases
customer_purchases = df.groupby('Customer_Name').size()

# Filter regular customers (let's say those who made more than 10 purchases)
regular_customers = customer_purchases[customer_purchases > 10]

# Plot the distribution of purchases made by regular customers
plt.figure(figsize=(10, 6))
regular_customers.hist(bins=20, color='skyblue')
plt.xlabel('Number of Purchases')
plt.ylabel('Number of Customers')
plt.title('Distribution of Purchases by Regular Customers')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Read the dataset
df = pd.read_csv('/content/superstore.csv', encoding="latin1")

# Step 2: Group data by customer and product and calculate total sales
customer_product_sales = df.groupby(['Customer_ID', 'Product_ID'])['Sales'].sum().reset_index()

# Step 3: Find top-selling product for each customer
top_selling_product_per_customer = customer_product_sales.loc[customer_product_sales.groupby('Customer_ID')['Sales'].idxmax()]

# Step 4: Visualize the top-selling product for each customer as a stacked bar plot
plt.figure(figsize=(12, 8))
sns.barplot(data=top_selling_product_per_customer, x='Customer_ID', y='Sales', hue='Product_ID', dodge=False)
plt.title('Top Selling Product for Each Customer')
plt.xlabel('Customer ID')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the dataset
df = pd.read_csv('/content/superstore.csv', encoding="latin1")

# Step 2: Extract year from Order_Date column
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Year'] = df['Order_Date'].dt.year

# Step 3: Group data by year and product and calculate total sales
yearly_product_sales = df.groupby(['Year', 'Product_ID'])['Sales'].sum().reset_index()

# Step 4: Find top-selling product for each year
top_selling_product_per_year = yearly_product_sales.loc[yearly_product_sales.groupby('Year')['Sales'].idxmax()]

# Step 5: Prepare data for grouped bar chart
pivot_table = top_selling_product_per_year.pivot(index='Year', columns='Product_ID', values='Sales').fillna(0)

# Step 6: Plot grouped bar chart
pivot_table.plot(kind='bar', stacked=False, figsize=(12, 8))
plt.title('Year-wise Top Selling Products')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend(title='Product ID', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset (replace 'dataset.csv' with your actual file path)
df = pd.read_csv('/content/superstore.csv', encoding="latin1")

# Convert Order_Date column to datetime format
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

# Extract year from Order_Date
df['Year'] = df['Order_Date'].dt.year

# Group by Year and sum up the sales
yearly_sales = df.groupby('Year')['Sales'].sum()

# Find the top selling year
top_selling_year = yearly_sales.idxmax()

# Plotting
plt.figure(figsize=(10, 6))
yearly_sales.plot(kind='bar', color='skyblue')
plt.title('Top Selling Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.axvline(x=top_selling_year, color='red', linestyle='--', label='Top Selling Year: ' + str(top_selling_year))
plt.legend()
plt.tight_layout()
plt.show()
```
